```python
import os
import pandas as pd
```

```python
gse = "GSE185557"
path = f"./sources/{gse}/"
filename = os.listdir(path)[0]
cnt_mtx = pd.read_csv(f"{path}{filename}")
```

```python
genes_traduction = pd.read_csv('./helper_data/ensembl_codes.csv', sep=',', header=None)
genes_traduction = dict(zip(genes_traduction.loc[:, 0], genes_traduction.loc[:, 1]))
```

```python
c=0
for indx, ensembl in enumerate(cnt_mtx['ID_REF']):
    try:
        cnt_mtx.loc[indx, "gene"] = genes_traduction[ensembl]
    except KeyError:
        c += 1
        cnt_mtx.loc[indx, "gene"] = ensembl
```

```python
c
```

```python
cnt_mtx.index = cnt_mtx["gene"]
cnt_mtx.drop("gene", axis=1, inplace=True)
cnt_mtx.drop("ID_REF", axis=1, inplace=True)
```
