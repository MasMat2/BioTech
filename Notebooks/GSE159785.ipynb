```python
import pandas as pd
import os
```

```python
datas_name = "GSE159785"
folder_path = f"../data_sources/New/{datas_name}/"
fn = f"{folder_path}/{os.listdir(folder_path)[0]}"
data = pd.read_csv(fn, sep="\t")
```

```python
persons = sorted(list(set(list(map(lambda x: x.split("_ROI")[0], data.columns)))))
len(persons)
persons
```

```python
from collections import defaultdict
persons_counter = defaultdict(lambda: 0)
```

```python
for column in data.columns:
    for person in persons:
        if person in column:
            persons_counter[person] += 1
```
