```python
import pandas as pd
import re
sources_path = '../0sources'
clean_data_path = '../2clean_data'
```

```python
with open(f"{sources_path}/GSE147507_family.soft", "r") as f:
    samples = f.read().split("^SAMPLE = ")
```

```python
samples[1]
```

```python
samples = samples[1:]
samples_dict = dict()
for sample in samples:
    name = sample.split('\n')[0]
    samples_dict[name] = dict()
    fields = ['tissue', 'cell line', 'cell type', 'tissue', 'treatment', 'subject status', 'time point', 'time after treatment']
    for field in fields:
        try:
            samples_dict[name][f"{field.replace(' ', '_')}"] = sample.split(f"{field}: ")[1].split("\n")[0]
        except:
#             print(field, name)
            pass
    samples_dict[name][f"title"] = sample.split(f"!Sample_title = ")[1].split("\n")[0]
        
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df = pd.DataFrame(samples_dict)
```

```python
df
```

```python
study_samples = df.loc[:, df.columns[(df.loc['cell_type'] == 'Lung Biopsy')]]
```

```python
study_samples
```

```python
study_samples = study_samples.drop(['cell_line', 'time_point', 'time_after_treatment', 'tissue'])
```

```python
study_samples
```

```python
for sample in study_samples:
    study_samples[sample]['treatment'] = 'COVID' if 'COVID' in study_samples[sample]['subject_status'] else 'Control' 
study_samples.rename(index={'treatment':'group'}, inplace=True)
```

```python
study_samples = study_samples.transpose()
```

```python
study_samples['GSM'] = study_samples.index
```

```python
study_samples.index = study_samples['title']
study_samples.index.name = None
```

```python
study_samples = study_samples.drop('title', axis=1)
study_samples
```

```python
study_samples.to_csv(f"{clean_data_path}/colData.GSE147507.csv")
```
