```python
import pandas as pd
sources_path = '../0sources'
clean_data_path = '../2clean_data/'
```

```python
df = pd.read_csv(f"{sources_path}/GSE185557.csv")
df = df.rename(columns={"ID_REF": ""})
df
```

```python
MB_df = df[df.columns.drop(list(df.filter(regex='CB')))]
CB_df = df[df.columns.drop(list(df.filter(regex='MB')))]
```

```python
line = pd.DataFrame(dict(map(lambda col: (col, 'Control' if col.split('-') != [''] and col.split('-')[-2]=='C' else 'Infected'), MB_df.columns[0:])), index=[0])
MB_df = pd.concat([line, MB_df]).reset_index(drop=True)
```

```python
line = pd.DataFrame(dict(map(lambda col: (col, 'Control' if col.split('-') != [''] and col.split('-')[-2]=='C' else 'Infected'), CB_df.columns[0:])), index=[0])
CB_df = pd.concat([line, CB_df]).reset_index(drop=True)
```

```python
MB_df
```

```python


MB_df.to_csv(f"{clean_data_path}/GSE185557.MB.csv", index=False)
CB_df.to_csv(f"{clean_data_path}/GSE185557.CB.csv", index=False)
```
