```python
import pandas as pd
import os
```

```python
#IMPORTAMOS LOS GENES QUE NOS INTERESAN

genes1 = pd.read_csv('../Genes/genes-metabolismo.csv', header = None)
genes2 = pd.read_csv('../Genes/genes-metabolismo.csv', header = None)
genes3 = pd.read_excel('../Genes/genes2Nuevo.xlsx',header = None)
genes = set(genes1[0]).union(set(genes2[0])).union(genes3[0])
```

```python
datas_name = "GSE175779"
folder_path = f"../.data_sources/New/{datas_name}/"
fn = f"{folder_path}/{os.listdir(folder_path)[0]}"
data = pd.read_csv(fn, sep="\t")
data
```

```python
genes_file = "ensembl_codes.csv"
fn = f"../Genes/{genes_file}"
genes_traduction = pd.read_csv(fn, sep=",",header=None)
```

```python
genes_traduction = dict(zip(genes_traduction.loc[:, 0], genes_traduction.loc[:, 1]))
```

```python
for indx, ensembl in enumerate(data["ENSEMBL"]):
    try:
        data.loc[indx, "gene"] = genes_traduction[ensembl]
    except KeyError:
        pass
```

```python
data
```

```python
# data.rename(columns={"ENSEMBL":"gene"}, inplace=True)
data.index = data["gene"]
data.drop("gene", axis=1, inplace=True)
```

```python
data
```

```python
for indx, dat in enumerate([data.columns[1+9*x:1+9*(x+1)] for x in range(4)]):
    data[f"person {indx+1} uninfected"] = data.loc[: ,dat[:5]].mean(axis=1)
    data[f"person {indx+1} infected"] = data.loc[: ,dat[5:]].mean(axis=1)
```

```python
means = data.filter(like="person")
means
```

```python
means = data.filter(like="person")
for GAPDH,Columna in zip(list(means.loc['GAPDH']),list(means)) :
    data[Columna.replace("mean", "")+" norm"] = data[Columna]/GAPDH
```

```python
output_dir = f"../fold-change/{datas_name}"
os.makedirs(output_dir, exist_ok=True)
for indx, gene in enumerate([genes1, genes2, genes3]):
    fold_change_data.loc[list(gene.loc[:,0])].to_csv(f"{output_dir}/fold_change_genes_{indx+1}_{datas_name}.csv")

output_dir = f"../Resultados/{datas_name}"
os.makedirs(output_dir, exist_ok=True)
data.to_csv(f'{output_dir}/Resultados_Analisis_{datas_name}.csv')
```
